*Exploratory Data analysis*

```python
plt.figure(figsize=(10,6))
sns.heatmap(data=weather.corr(),annot=True)
plt.show()
```

```python
weather_monthly = weather.resample('MS').mean()
weather_monthly.head()
```

```python
weather_monthly.drop(['Temperature (C)','Wind Speed (km/h)','Wind Bearing (degrees)','Visibility (km)','Pressure (millibars)'],axis=1,inplace=True)
```

```python
weather_monthly.head()
```

```python
def label_color(month):
    if month == 1:
        return 'January','blue'
    elif month == 2:
        return 'February','green'
    elif month == 3:
        return 'March','orange'
    elif month == 4:
        return 'April','yellow'
    elif month == 5:
        return 'May','red'
    elif month == 6:
        return 'June','violet'
    elif month == 7:
        return 'July','purple'
    elif month == 8:
        return 'August','black'
    elif month == 9:
        return 'September','brown'
    elif month == 10:
        return 'October','darkblue'
    elif month == 11:
        return 'November','grey'
    else:
        return 'December','pink'

def plot_month(month, data):
    label, color = label_color(month)
    mdata = data[data.index.month == month]
    sns.lineplot(data=mdata,
                 label=label,
                 color=color,
                 marker='o')
    
def sns_plot(title, data):
    plt.figure(figsize=(14,8))
    plt.title(title)
    plt.xlabel('YEAR')
    for i in range(1,13):
        plot_month(i,data)
    plt.show()
```

```python
title = 'Plot of Apparent Temperature (C) - Month-wise' 
sns_plot(title, weather_monthly.iloc[:,0])
```

```python
title = 'Plot of Humidity- Month-wise' 
sns_plot(title, weather_monthly.iloc[:,1])
```

```python
sns.distplot(weather_monthly['Apparent Temperature (C)'])
plt.show()
sns.distplot(weather_monthly['Humidity'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Variation of Apparent Temperature and Humidity')
sns.lineplot(data=weather_monthly)
plt.show()
```

```python
#Humidity and Apparent Temperature monthwise 
def line_plot(data):
    for i in range(1,13):
        mon = data[data.index.month==i]
        plt.figure(figsize=(10,3))
        plt.plot(mon['Humidity'],label= 'Humidity',color='blue',linestyle='dashed')
        plt.plot(mon['Apparent Temperature (C)'],label = 'Apparendt Temperature (C)',color='red')
        plt.title('Apparante Temperature and Humidity variation Monthwise')
        plt.legend(loc=0)
        
```

```python
line_plot(weather_monthly)
```

```python
from scipy.stats import ttest_rel
data1 = weather_monthly['Apparent Temperature (C)']
data2 = weather_monthly['Humidity']
test = ttest_rel(data1,data2)
stat , p = test[0],test[1]
print('stat=%.3f, p=%.3f' % (stat, p))
```

```python
if p > 0.05:
    print('Accept Null Hypothesis')
else:
    print('Reject Null Hypothesis')
```
